```
%r

library(lattice)
histogram(~Cholesterol | Gender, data=data, layout=c(1,2))

# Compute the mean cholesterol value for each Gender;
male.cholesterol <- subset(data$Cholesterol,data$Gender=='Male');
female.cholesterol <- subset(data$Cholesterol,data$Gender=='Female');
mean_male_col <- mean(male.cholesterol)
mean_fmal_col <-mean(female.cholesterol)
cat("Male Col:",mean_male_col,"\nFemale Col:", mean_fmal_col)

# Use these values to interpret and validate the ANOVA model;
cat("\n\nDifference in means = beta coeffieicnt:",mean_male_col - mean_fmal_col)
```
