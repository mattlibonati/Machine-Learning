### Confusion Matrix: 
Also known as an 'error matrix'; used to describe the performance of a classification model on test data where the true values are known.

<b> Libraries </b>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<b> Setting up Confusion Matrix </b>

```python
confusion_matrix = np.ceil(np.random.rand(2,2)*100000)
print('Confusion Matrix')
display(pd.DataFrame(confusion_matrix,columns=['N_predict','p_predict']).rename(index={0:'N_actual',1:'P_actual'}))  
```

<b> Precision and Recall </b>

```python
TN = confusion_matrix[0][0]
FN = confusion_matrix[1][0]
FP = confusion_matrix[0][1]
TP = confusion_matrix[1][1]

precision           = round((TP)/(TP+FP),2)
recall              = round((TP)/(TP+FN),2)
specificity         = round((TN)/(TN+FP),2) 
false_positive_rate = round((FP)/(FP+TN),2)

print(f'Precision = (TP)/(TP+FP) = ({TP})/({TP}+{FP}) = {precision}')
print(f'Recall (Sensitivity) = (TP)/(TP+FN) = ({TP})/({TP}+{FN}) = {recall}')
print(f'Specificity = (TN)/(TN+FP) = ({TN})/({TN}+{FP}) = {specificity}')
print(f'False Positive Rate = (FP)/(FP+TN) = ({FP})/({FP}+{TN}) = {false_positive_rate}')
```

<b> Plot Data </b>

```python
# line arrays
true_positive_rate_p  = [0, recall, 1]
true_negative_rate_p  = [0, specificity, 1]
precision_p           = [0, precision, 1]
false_positive_rate_p = [0, false_positive_rate, 1]

# plots
fig, axes = plt.subplots(nrows=1,ncols=2)
fig.set_figheight(5)
fig.set_figwidth(15)

# ROC Curve
axes[0].plot(false_positive_rate_p,true_positive_rate_p,color='r',label='ROC')
axes[0].plot([0,1],[0,1],color='b',label='Baseline')
axes[0].set_title('Receiver Operating Characteristic (ROC) Curve')
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].legend(loc='upper left')

# Precision Vs Re
axes[1].plot(false_positive_rate_p, precision_p, color='red',label='Prec v Recall')
axes[1].set_title('Precision vs Recall')
axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].legend(loc='upper left')

plt.show()
```

<b> Bayes Theorem: </b> <br>
Probability of an event, based on prior knowledge of conditions that might be related to the event.

```python
P_A   = (TP + FN) / confusion_matrix.sum()
P_B   = (TP + FP) / confusion_matrix.sum()
bayes = (recall * P_A) / P_B
print(f'Bayes: {bayes:.2f}')
```
