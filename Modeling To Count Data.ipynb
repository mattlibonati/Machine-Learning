<b> 4. Model Identification: </b> <br> <br>
Use backwards variable selection on the full data frame (of valid predictor variables!) for each of the count regression models. By using backwards variable selection your upper limit is the full model and your lower limit is the intercept model.

```
%r
library(MASS)
```

(1) Poisson regression using the glm() function

```
%r
# Define the upper model as the FULL model
upper.glm <- glm(ofp ~ .,data=train_use, family = poisson);

# Define the lower model as the Intercept model
lower.glm <- glm(ofp ~ 1,data=train_use, family = poisson);

backward_1.glm <- stepAIC(object=upper.glm, direction=c('backward'));
print(summary(backward_1.glm))
```

(2) Poisson regression with dispersion using the glm() function

```
%r
# Define the upper model as the FULL model
upper.glm <- glm(ofp ~ .,data=train_use, family = poisson(link = "log"));

# Define the lower model as the Intercept model
lower.glm <- glm(ofp ~ 1,data=train_use, family = poisson(link = "log"));

backward_2.glm <- stepAIC(object=upper.glm, direction=c('backward'));
print(summary(backward_2.glm))
```

```
%r
gof.ins <- function(model, df){
  
  abs_vals <- abs(df$ofp-predict(model, df))/df$ofp
  abs_vals[which(!is.finite(abs_vals))] <- 0
  
  cat('AIC:',AIC(model), '\nBIC:',BIC(model), '\nMSE:', mean(model$residuals^2), '\nMAE:', mae(df$ofp, predict(model, df)), '\nMAPE:', mean(abs_vals) * 100, '\n')
  
  backward.testPCT <- abs(df$ofp-predict(model, df))

  backward.testPredictionGrade <- ifelse(backward.testPCT<=2,'Grade 1: [0,2]',
  ifelse(backward.testPCT<=4,'Grade 2: (2,4]',
  ifelse(backward.testPCT<=6,'Grade 3: (4,6]',
  'Grade 4: (6+]')))

  backward.testTable <-table(backward.testPredictionGrade)
  print(backward.testTable/sum(backward.testTable))
  }

gof.oos <- function(model, df){

  abs_vals <- abs(df$ofp-predict(model, df))/df$ofp
  abs_vals[which(!is.finite(abs_vals))] <- 0
  
  cat('MSE:', mean((df$ofp - predict(model, df)) ^ 2), '\nMAE:', mae(df$ofp, predict(model, df)), '\nMAPE:', mean(abs_vals) * 100, '\n')
  
  backward.testPCT <- abs(df$ofp-predict(model, df))

  backward.testPredictionGrade <- ifelse(backward.testPCT<=2,'Grade 1: [0,2]',
  ifelse(backward.testPCT<=4,'Grade 2: (2,4]',
  ifelse(backward.testPCT<=6,'Grade 3: (4,6]',
  'Grade 4: (6+]')))
  
  backward.testTable <-table(backward.testPredictionGrade)
  print(backward.testTable/sum(backward.testTable))
  
  }
```

```
%r
print('Model 1: Poisson Regression:')
print(backward_1.glm)
print('\nINS')
print(gof.ins(backward_1.glm,train_use))
print('OOS')
print(gof.oos(backward_1.glm,test_use))
```

Let’s apply some context to our problem and see how well it might work in a business application. Suppose that the actual prediction of the count regressions does not matter. Instead we want to know if we can classify people into segments based on their expected number of physician office visits.

- Segment #1 = 0-5 physician office visits
- Segment #2 = 6-10 physician office visits
- Segment #3 = 11+ physician office visits

Round the predicted values if needed. Produce the in-sample and out-of-sample confusion matrices for each model. Make a summary table of the overall classification accuracy for each model in sample and out-of-sample. The business strategy requires that the models be 80% accurate out-of sample in order for the strategy to work properly. Do your models meet this accuracy threshold?

```
%r
library(dplyr)
```

```
%r

confusion_view <- function(model, df){

  confusion <- data.frame(df$ofp,predict(model, df))
  names(confusion) <- c('actual','prediction')

  confusion$actual_cat <- 0

  confusion <- confusion %>%
    mutate(actual_cat = case_when(
      actual <= 5 ~ '0-5',
      actual <= 10 ~ '6-10',
      actual > 10 ~ '11+',
      ))

  confusion <- confusion %>%
    mutate(predict_cat = case_when(
      prediction <= 5 ~ '0-5',
      prediction <= 10 ~ '6-10',
      prediction > 10 ~ '11+',
      ))

  confusion <- confusion %>%
    mutate(result = case_when(
      prediction <= 5 ~ '0-5',
      prediction <= 10 ~ '6-10',
      prediction > 10 ~ '11+',
      ))

  dcast(data = confusion, formula = actual_cat ~ predict_cat, fun.aggregate = length, value.var = "actual")
  
  }
```

```
%r
print('Model 1: Poisson Regression:')
print(confusion_view(backward_1.glm,test_use))
print(confusion_view(backward_1.glm,train_use))
```
